```python
# Libs
import torch
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
```

```python
# Fleurs dataset
from datasets import load_dataset
fleurs = load_dataset("google/fleurs", "en_us", split="train", trust_remote_code=True)
```

```python
from IndicTransToolkit.processor import IndicProcessor
```

```python
# IndicTrans2 
src_lang, tgt_lang = "eng_Latn", "hin_Deva"
model_name = "ai4bharat/indictrans2-en-indic-1B"
tokenizer = AutoTokenizer.from_pretrained(model_name, trust_remote_code=True)
ip = IndicProcessor(inference=True)
batch_size = 10
all_vectors = []
model = AutoModelForSeq2SeqLM.from_pretrained(
    model_name,
    trust_remote_code=True,
    torch_dtype=torch.float16, # performance might slightly vary for bfloat16
    attn_implementation="flash_attention_2"
).to(DEVICE)
```

```python
for i in range(0, len(fleurs), batch_size):
  batch_sentences = fleurs['transcription'][i:i + batch_size]
  batch_ids = fleurs['id'][i:i + batch_size]


  # Generate translations using the model
  batch = ip.preprocess_batch(
      batch_sentences,
      src_lang=src_lang,
      tgt_lang=tgt_lang,
  )

  # Tokenize the sentences and generate input encodings
  inputs = tokenizer(
      batch,
      truncation=True,
      padding="longest",
      return_tensors="pt",
      return_attention_mask=True,
  ).to(DEVICE)


  # Step 3: Run decoder manually using encoder outputs
  with torch.no_grad():
      encoder_outputs = model.model.encoder(
          input_ids = inputs["input_ids"],
          attention_mask = inputs["attention_mask"]
      )
  encoder_hidden_states = encoder_outputs.last_hidden_state

  for idx_in_batch, example_id in enumerate(batch_ids):
    encoded_vector = encoder_hidden_states[idx_in_batch].cpu()  # shape: (T_i, H)
    example_dict = {
        "id": example_id,
        "encoded_vector": encoded_vector  # tensor of shape (T_i, 1024)
    }
    all_vectors.append(example_dict)
      
  print(f"Processed batch with ID: {i}")
```
